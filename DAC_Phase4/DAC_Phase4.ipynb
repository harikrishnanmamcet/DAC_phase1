```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (replace 'your_dataset.csv' with the actual file path)
data = pd.read_csv('/content/20140711 (1).csv')

# Basic data exploration
print(data.head())  # Display the first few rows of the dataset
print(data.info())  # Get information about the dataset, e.g., data types

# Visualization: On-Time Performance Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='WeekBeginning', y='NumberOfBoardings', data=data)
plt.xlabel('Week Beginning')
plt.ylabel('Number of Boardings')
plt.title('On-Time Performance Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Visualization: Distribution of Number of Boardings
plt.figure(figsize=(10, 6))
sns.histplot(data['NumberOfBoardings'], bins=30, kde=True)
plt.xlabel('Number of Boardings')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Boardings')
plt.grid(True)
plt.show()
```
