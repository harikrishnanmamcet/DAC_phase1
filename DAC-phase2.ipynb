```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the provided data
data = pd.DataFrame({
    'TripID': [23631, 23631, 23632, 23633, 23633, 23634, 23634, 23634, 23634],
    'RouteID': [100, 100, 100, 100, 100, 100, 100, 100, 100],
    'StopID': [14156, 14144, 14132, 12266, 14147, 13907, 14132, 13335, 13875],
    'StopName': ['181 Cross Rd', '177 Cross Rd', '175 Cross Rd', 'Zone A Arndale Interchange', '178 Cross Rd', '9A  Marion Rd', '175 Cross Rd', '9A  Holbrooks Rd', '9  Marion Rd'],
    'WeekBeginning': ['30-06-2013 00:00', '30-06-2013 00:00', '30-06-2013 00:00', '30-06-2013 00:00', '30-06-2013 00:00', '30-06-2013 00:00', '30-06-2013 00:00', '30-06-2013 00:00', '30-06-2013 00:00'],
    'NumberOfBoardings': [1, 1, 1, 2, 1, 1, 1, 1, 1]
})

# Group the data by 'StopName' and sum the 'NumberOfBoardings'
boarding_counts = data.groupby('StopName')['NumberOfBoardings'].sum()

# Create a bar chart
plt.figure(figsize=(10, 6))
boarding_counts.plot(kind='bar', color='red')
plt.title('Number of Boardings at Bus Stops')
plt.xlabel('Bus Stop')
plt.ylabel('Number of Boardings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
